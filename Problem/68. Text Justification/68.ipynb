# Text Justification

**Problem**:
Given an array of strings `words` and an integer `maxWidth`, format the text such that each line has exactly `maxWidth` characters, fully justified (both left and right).

- Pack as many words as you can in each line using a greedy approach.
- Distribute extra spaces between words as evenly as possible. If the spaces don't divide evenly, the left slots should get more spaces than the right ones.
- The last line should be left-justified, with no extra space between words.

**Examples**:

**Input**:
`words = ["This", "is", "an", "example", "of", "text", "justification."]`, `maxWidth = 16`

**Output**:

[

"This is an      ",

"example of text ",

"justification.  "

]

&nbsp;

**Input**:
`words = ["What","must","be","acknowledgment","shall","be"]`, `maxWidth = 16`

**Output**:
[
"What must be   ",

"acknowledgment ",

"shall be       "
]


**Explanation**:
The last line is left-justified as "shall be    ", not "shall     be". The second line is also left-justified because it contains only one word.


**Constraints**:
- A word is a character sequence consisting of non-space characters only.
- Each word's length is `> 0` and `<= maxWidth`.
- `1 <= words.length, maxWidth <= 1000`
