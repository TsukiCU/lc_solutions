# Sort a Nearly Sorted (K-Sorted) Array

**Problem Description**:
A nearly sorted array is defined as an array where each element is at most `k` positions away from its sorted position. The task is to choose an appropriate sorting algorithm that efficiently sorts such an array.

**Example 1**:
- Input: `arr = [6, 5, 3, 2, 8, 10, 9]`, `k = 3`
- Output: `[2, 3, 5, 6, 8, 9, 10]`
- Explanation: After sorting, each element in the array is within 3 positions of its sorted position.

**Example 2**:
- Input: `arr = [10, 9, 8, 7, 4, 70, 60, 50]`, `k = 4`
- Output: `[4, 7, 8, 9, 10, 50, 60, 70]`
- Explanation: The array is almost sorted, as each element is no more than 4 positions away from its sorted order.

**Approach**:
A suitable algorithm for this scenario is a modified insertion sort or using a min-heap, which can sort the array in O(n log k) time complexity.


```python
import heapq

class Solution:
    '''
        Dual pointers, one points to the next element to be pushed
        and one to the next element to be removed.
    '''
    def nearlysorted(self, nums, k):
        k = min(len(nums), k)
        index = k+1

        heap = nums[:index]
        heapq.heapify(heap)

        for i in range(len(nums)-k-1):
            heapq.heappush(heap, nums[index])
            index += 1
            nums[i] = heapq.heappop(heap)

        while heap:
            nums[i+1] = heapq.heappop(heap)
            i += 1


nums = [10, 9, 8, 7, 4, 70, 60, 50]
k = 4
s = Solution()
s.nearlysorted(nums, k)
nums
```
